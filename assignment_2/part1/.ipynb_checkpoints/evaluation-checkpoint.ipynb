```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
import random
```

```python
file_path = Path.cwd() / 'output'
model_types = ["RNN", "LSTM"]

file = file_path / ("results_" + model_types[0] + ".csv")
df_rnn = pd.read_csv(file, delimiter=';', header=0)

file = file_path / ("results_" + model_types[1] + ".csv")
df_lstm = pd.read_csv(file, delimiter=';', header=0)

print(df_rnn.head())
```

```python
df_rnn.iloc[0]
```

```python
df_rnn = df_rnn.drop(columns=['input_dim','num_hidden','lr','train_steps','batch_size'])
df_lstm = df_lstm.drop(columns=['input_dim','num_hidden','lr','train_steps','batch_size'])
```

```python
df_rnn = df_rnn.groupby(['seq_length']).mean()
df_lstm = df_lstm.groupby(['seq_length']).mean()
```

```python
df_rnn[:]['mean acc']
```

```python
df_rnn = df_rnn.drop(df_rnn.index[-1])
```

```python
x = range(5, 71, 5)
```

```python
df_rnn['mean acc']
```

```python
fig = plt.figure()  # an empty figure with no axes
fig.suptitle('No axes on this figure')  # Add a title so we know which it is

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,5))  # a figure with a 2x2 grid of Axes

ax1.plot(df_rnn['mean acc'], label='RNN')
ax1.plot(df_lstm['mean acc'], label='LSTM')

ax1.set_xlabel('Sequence Length', size=12)
ax1.set_ylabel('Accuracy', size=12)
ax1.grid(True)
ax1.legend()

#loss
ax2.plot(df_rnn['mean loss'], label='RNN')
ax2.plot(df_lstm['mean loss'], label='RNN')

ax2.set_xlabel('Sequence Length', size=12)
ax2.set_ylabel('Loss', size=12)
ax2.grid(True)
ax2.legend()

#plt.title("Model Performance for Palindrome Task")
plt.suptitle("Model Performance over Palindrome Length", size=16)
plt.show()
```
