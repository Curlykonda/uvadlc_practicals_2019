```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
import random
```

```python
file_path = Path.cwd() / 'output'
books = ['alice', 'grimm', 'democracy']
results = {}
for b in books:
    file = file_path / (b + "_results.npy")
    results[b] = np.load(file)
```

```python
len(results['alice']) #step, acc, loss
```

```python
steps, acc, loss = list(zip(*results['alice']))
steps = [int(s) for s in steps]
plt.plot(steps, acc)
```

```python
interval = steps[1] - steps[0]
interval
```

```python
fig = plt.figure()  # an empty figure with no axes
fig.suptitle('No axes on this figure')  # Add a title so we know which it is

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,5))  # a figure with a 2x2 grid of Axes

for b in books:
    steps, acc, loss = list(zip(*results[b]))
    ax1.plot(steps, acc, label=b)
    #loss
    ax2.plot(steps, loss, label=b)
    print(f"{b}: best acc {np.max(acc):.3f} at step {np.argmax(acc)*interval}")
    print(f"{b}: best loss {np.min(loss):.3f} at step {np.argmin(loss)*interval}")

    
    
ax1.set_xlabel('Training Step', size=12)
ax1.set_ylabel('Accuracy', size=12)
ax1.grid(True)
ax1.legend()
ax2.set_xlabel('Training Step', size=12)
ax2.set_ylabel('Loss', size=12)
ax2.grid(True)
ax2.legend()

#plt.title("Model Performance for Palindrome Task")
plt.suptitle("LSTM Training Performance for different books", size=16)
plt.show()
```
