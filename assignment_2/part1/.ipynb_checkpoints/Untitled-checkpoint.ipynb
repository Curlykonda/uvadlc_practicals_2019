```python
import itertools
import copy
import numpy as np
import random

jungs = [(f"J{x}") for x in range(1,7)]
maedels = [(f"M{x}") for x in range(1,10)]
team_a, team_b = [], []
i = 0
while jungs:
    if i % 2:
        team_a.append(jungs.pop())
    else:
        team_b.append(jungs.pop())
    i+=1
while maedels:
    if i % 2:
        team_a.append(maedels.pop())
    else:
        team_b.append(maedels.pop())
    i+=1

thresh = 4
num_matches = 30
game_plan = []

while len(game_plan) < num_matches:
    teams = {'a': copy.deepcopy(team_a), 'b': copy.deepcopy(team_b)}
    p_count = {p: 0 for p in [*team_a, *team_b]}
    matchups = set(itertools.product(*[teams['a'], teams['b']]))
    matchups = matchups - set(game_plan)
    stop=False
    while not stop:
        #print(len(matchups))
        #try:
        okay=False
        while not okay:
            rnd_m = random.choice(list(matchups))
            if rnd_m not in game_plan:
                game_plan.append(rnd_m)
                matchups.remove(rnd_m)
                okay=True
        #print(f"{rnd_m}")
        #print(len(game_plan))
        for p in rnd_m:
            #print(f"{p}")
            i=0
            p_count[p] += 1
            if p_count[p] == thresh:
                p_count.pop(p, None)
                if p in teams['a']:
                    teams['a'].remove(p)
                elif p in teams['b']:
                    teams['b'].remove(p)

        if len(game_plan) == num_matches:
            print("Done")
            break
                #print(f"Remove:{p}")

                #print(len(p_count))
        matchups = set(itertools.product(*[teams['a'], teams['b']]))
        matchups = matchups - set(game_plan)
        #print(len(matchups))
        if len(matchups) == 0:
            stop=True
        elif len(p_count.keys()) < 2:
            stop=True
            
    print(f"Gp:{len(game_plan)}")
    
print(team_a)
print(team_b)
random.shuffle(game_plan)
for i, g in enumerate(game_plan):
    print(f"{g[0]}-{g[1]}")
```
